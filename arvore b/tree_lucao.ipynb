```python
class Node():
    def __init__(self, leaf=True):
        self.keys = []
        self.children = []
        self.leaf = leaf
    def is_full(self):
        return len(self.keys) == 3
class BTree():
    def __init__(self):
        self.root = Node(True)
    def insert(self, value):
        """
        Arguments:
            value (int): value to be added to the tree
        """
        # first step is take the root and check whether it's full or not
        active_node = self.root
        import  pdb
        pdb.set_trace()
        # if the node is full
        if active_node.is_full():
            # we determine the mid point of the node to be split and divide its keys and children between
            # new nodes
            mid_point = len(active_node.keys) // 2
            mid_key = active_node.keys[mid_point]
            # we first create a new parent node
            new_parent_node = Node()
            # we append the current root as child of the new parent node
            new_parent_node.keys.append(mid_key)
            new_parent_node.children.append(active_node)
            new_parent_node.children.sort()
            new_parent_node.leaf = False
            self.root = new_parent_node
            # we create a new child
            new_child_node = Node()
            # and reassign the values
            new_child_node.keys = active_node.keys[mid_point+1:]
            new_child_node.children = active_node.children[mid_point+1:]
            # clonando 
            active_node.keys = active_node.keys[:mid_point]
            # for the children cloning we have to clone :mid_point + 1
            active_node.children = active_node.children[:mid_point+1]
            new_parent_node.children.append(new_child_node)
            if len(new_child_node.children) > 0:
                new_child_node.leaf = False
            if value > mid_key:
                active_node = new_child_node
        while not active_node.leaf:
            key_indexer = len(active_node.keys) - 1
            while active_node.keys[key_indexer] > value:
                key_indexer -= 1
            if active_node.keys[key_indexer] < value :
                key_indexer += 1
            next_active_node = active_node.children[key_indexer]
            if next_active_node.is_full():
                # we first define the new mid point
                mid_point = len(next_active_node.keys) // 2
                mid_key = next_active_node.keys[mid_point]
                # we create a new Node
                new_child_node = Node()
                # we append it as child of the active node children
                active_node.children.append(new_child_node)
                # we promote the mid key of the node being split to the parent node
                active_node.keys.append(mid_key)
                # and sort it
                active_node.keys.sort()
                # we now then split the keys between the currently being split node 
                # and the new child one
                new_child_node.keys = next_active_node.keys[mid_point+1:]
                new_child_node.children = next_active_node.children[mid_point+1:]
                next_active_node.keys = next_active_node.keys[:mid_point]
                next_active_node.children = next_active_node.children[:mid_point]
                if value > mid_key:
                    active_node = new_child_node
                else:
                    active_node = next_active_node
            else:
                active_node = next_active_node
        active_node.keys.append(value)
        active_node.keys.sort()
```
